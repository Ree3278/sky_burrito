```python
from simulation.rl_training import CurriculumCallback, PPO, create_environment, evaluate_model

device = "cpu"
fleet_size = 30
phase = 3
resume_path = "models/fleet_20/ppo_fleet_20_phase_3.zip"
curriculum = CurriculumCallback(phase = phase)
env = create_environment(fleet_size, phase, curriculum)
model = PPO.load(str(resume_path), env=env, device=device)
print(f"\nTesting trained model...")
test_reward = evaluate_model(model, env, num_episodes=3)
print(f"✓ Average test reward: {test_reward:.2f}")
env.close()
```
